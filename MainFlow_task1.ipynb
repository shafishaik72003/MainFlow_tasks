```python
!pip install pandas matplotlib seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv('student-mat.csv', sep=';')

# Display first few rows
df.head()
# Check for missing values
df.isnull().sum()
df.drop_duplicates(inplace=True)

average_g3 = df['G3'].mean()
print("Average Final Grade (G3):", round(average_g3, 2))
high_scorers = df[df['G3'] > 15].shape[0]
print("Number of students scoring above 15:", high_scorers)
correlation = df[['studytime', 'G3']].corr()
print(correlation)
average_g3_by_gender = df.groupby('sex')['G3'].mean()
print(average_g3_by_gender)
plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], bins=10, kde=True)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("G3 Score")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['studytime'], y=df['G3'])
plt.title("Study Time vs. Final Grade")
plt.xlabel("Study Time (Hours per Week)")
plt.ylabel("G3 Score")
plt.show()
plt.figure(figsize=(6, 4))
sns.barplot(x=df['sex'], y=df['G3'])
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average G3 Score")
plt.show()
```
